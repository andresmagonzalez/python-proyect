```python
# 1. Pedimos los dos números enteros al usuario
num1 = int(input("Ingrese el primer número entero: "))
num2 = int(input("Ingrese el segundo número entero: "))

# 2. Nos aseguramos de saber cuál es el menor y cuál el mayor de los dos
# (Esto evita errores si el usuario ingresa los números al revés)
inicio = min(num1, num2)
fin = max(num1, num2)

# 3. Creamos una variable acumuladora para guardar la suma
suma_pares = 0

# 4. Recorremos los números en el rango seleccionado
# Sumamos +1 al 'fin' porque en range() el último número no se incluye
for numero in range(inicio, fin + 1):
    # Verificamos si el número es par (si el residuo de dividir entre 2 es 0)
    if numero % 2 == 0:
        suma_pares += numero  # Usamos el operador += que aprendiste antes

# 5. Mostramos el resultado final
print(f"La suma de los números pares entre {inicio} y {fin} es: {suma_pares}")
```

```python
# min(num1, num2) y max(num1, num2): Estas funciones identifican automáticamente cuál número es el más chico (para usarlo como punto de partida) y cuál es el más grande (para usarlo como límite). Así, si el usuario ingresa 10 y 2, el programa ordenará el rango de 2 a 10 sin problemas.

# range(inicio, fin + 1): La función range(a, b) genera números desde a hasta uno antes de b. Al sumarle +1 a la variable fin, nos aseguramos de que si el número final es par, también sea tomado en cuenta en la suma.

# El operador % (Módulo): Este operador calcula el residuo de una división. En Python, si haces numero % 2 == 0 y da verdadero, significa que el número es perfectamente divisible por 2, es decir, es un número par.
```

```python
# Ejercicio 2:
import math

print("--- Calculadora de Áreas ---")
print("Opciones: cuadrado, triangulo, circulo")
# Convertimos la respuesta a minúsculas con .lower() para evitar errores si escriben con mayúsculas
figura = input("¿De qué figura geométrica quieres calcular el área?: ").lower()

# 1. Caso para el Cuadrado
if figura == "cuadrado":
    lado = float(input("Ingresa la medida del lado del cuadrado: "))
    area = lado * lado  # O también: lado ** 2
    print(f"El área del cuadrado es: {area:.2f}")

# 2. Caso para el Triángulo
elif figura == "triangulo" or figura == "triángulo":
    base = float(input("Ingresa la base del triángulo: "))
    altura = float(input("Ingresa la altura del triángulo: "))
    area = (base * altura) / 2
    print(f"El área del triángulo es: {area:.2f}")

# 3. Caso para el Círculo
elif figura == "circulo" or figura == "círculo":
    radio = float(input("Ingresa el radio del círculo: "))
    # Usamos math.pi para el valor de Pi
    area = math.pi * (radio ** 2)
    print(f"El área del círculo es: {area:.2f}")

# 4. Caso por si el usuario escribe otra cosa
else:
    print("Figura no reconocida. Por favor, intenta de nuevo con 'cuadrado', 'triangulo' o 'circulo'.")
```

```python
#Que herramientas se utilizaron?
#import math: Esta librería viene incluida en Python. Al usar math.pi, el programa utiliza el valor preciso de Pi (3.141592...) en lugar de obligarte a escribir 3.14.

#.lower(): Si el usuario escribe "CUADRADO" o "Cuadrado", este método lo transforma automáticamente a "cuadrado". Así evitamos que el programa falle por culpa de las mayúsculas.

#float(input(...)): Usamos float en lugar de int porque las medidas de las figuras (como el radio de un círculo o la altura de un triángulo) perfectamente pueden tener decimales (ej. 5.5 cm).

#{area:.2f}: Dentro de la F-string, el :.2f le indica a Python que redondee el resultado a solo 2 decimales. Esto es especialmente útil con el círculo, ya que multiplicar por Pi suele dejar una fila enorme de números decimales.
```

```python
#Ejercicio 3:
import random

# 1. Generamos el número aleatorio entre 1 y 100
numero_secreto = random.randint(1, 100)

print("--- ¡Adivina el Número Secreto! ---")
print("He pensado un número entre 1 y 100. Intenta adivinarlo.\n")

# 2. Inicializamos una variable para la adivinanza del usuario
# La creamos en 0 para que sea diferente al número secreto y el ciclo pueda iniciar
adivinanza = 0
intentos = 0  # Llevaremos la cuenta de cuántos intentos le toma ganar

# 3. Iniciamos el ciclo while
# El ciclo se ejecutará MIENTRAS la adivinanza sea diferente (!=) al número secreto
while adivinanza != numero_secreto:
    adivinanza = int(input("Introduce tu número: "))
    intentos += 1  # Sumamos un intento
    
    # Damos pistas
    if adivinanza < numero_secreto:
        print("El número secreto es MAYOR.")
    elif adivinanza > numero_secreto:
        print("El número secreto es MENOR.")
    print("-" * 30)

# 4. El ciclo termina cuando adivinanza == numero_secreto
print(f"\n¡Felicidades! Has adivinado el número.")
print(f"El número era el {numero_secreto} y te tomó {intentos} intentos conseguirlo.")
```

```python
#Explicamos el uso del ciclo while, el cual ha sido utilizado como interruptor basado en la condicion:
#La condición adivinanza != numero_secreto: Significa "mientras la adivinanza sea diferente al número secreto, sigue preguntando".

    #El truco del inicio (adivinanza = 0): Al empezar el programa, necesitamos que Python entre al ciclo por primera vez. Como el número secreto está entre 1 y 100, asignar un 0 asegura que la condición de "ser diferentes" sea verdadera y el juego comience.

    #El final del ciclo: En el momento exacto en que el usuario ingresa el número correcto, la condición adivinanza != numero_secreto se vuelve falsa (porque ahora son iguales). En ese instante, Python rompe el ciclo automáticamente y pasa a la última línea para felicitar al jugador.
```

```python
#Ejercicio 4:
import time

# 1. Solicitamos la velocidad al usuario
velocidad = float(input("Ingresa la velocidad del pollo (metros por segundo): "))

# 2. Inicializamos la distancia y el tiempo
distancia_recorrida = 0
segundo = 0
meta = 100

print("\n¡Empieza la carrera!\n")

# 3. Ciclo while que se ejecuta segundo a segundo hasta llegar a la meta
while distancia_recorrida < meta:
    segundo += 1
    distancia_recorrida += velocidad
    
    # Nos aseguramos de que no dibuje más allá de la meta
    if distancia_recorrida > meta:
        distancia_recorrida = meta
        
    # Creamos el camino usando guiones (o espacios) según los metros recorridos
    camino = "-" * int(distancia_recorrida)
    
    # Mostramos el avance en la pantalla
    print(f"{camino}🐥 (Segundo {segundo}: {distancia_recorrida}m)")
    
    # Pausa de 1 segundo para simular el tiempo real (opcional, pero recomendado)
    time.sleep(1)

# 4. Mensaje final al cumplir la condición
print("\nel pollito ha llegado a la meta 🎉")
```
