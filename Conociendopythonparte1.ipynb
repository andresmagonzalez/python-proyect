```python
#1.- CALCULANDO EL SALARIO SEMANAL
# 1. Solicitar al usuario el monto de las ventas totales
# Usamos float() para permitir que se ingresen decimales (centavos)
ventas_totales = float(input("Ingrese el monto de las ventas totales de la semana: $"))

# 2. Definir las constantes del problema
salario_base = 500.0
porcentaje_comision = 0.10  # 10% expresado en decimal

# 3. Calcular la comisión y el salario total
comision = ventas_totales * porcentaje_comision
salario_total = salario_base + comision

# 4. Mostrar los resultados en pantalla con formato de dos decimales
print("\n--- Resumen de Pago ---")
print(f"Salario Base: ${salario_base:.2f}")
print(f"Comisión por ventas (10%): ${comision:.2f}")
print(f"-----------------------")
print(f"Salario Total Semanal: ${salario_total:.2f}")
```

```python
#2.- Bienvenido a la frutería virtual

# 1. Crear el diccionario vacío
catalogo_frutas = {}

# 2. Agregar algunas frutas con sus respectivos precios
catalogo_frutas['manzana'] = 1.50
catalogo_frutas['platano'] = 0.80
catalogo_frutas['naranja'] = 1.20
catalogo_frutas['fresa'] = 2.50

# 3. Pedir al usuario que ingrese el nombre de una fruta
# Usamos .lower() para que no importe si escriben con mayúsculas o minúsculas
fruta_buscada = input("¿Qué fruta estás buscando? ").lower()

# 4. Verificar si la fruta está en el diccionario y mostrar el resultado
if fruta_buscada in catalogo_frutas:
    precio = catalogo_frutas[fruta_buscada]
    print(f"¡Sí tenemos! La fruta '{fruta_buscada}' cuesta ${precio:.2f}")
else:
    print(f"Lo sentimos, la fruta '{fruta_buscada}' no se encuentra en nuestra tienda.")
    #lower(): Al usarlo en la entrada del usuario, si alguien escribe "Manzana" o "MANZANA", el programa lo convertirá a "manzana" internamente. Así coincidirá perfectamente con las claves que guardamos en el diccionario.
    #if fruta_buscada in catalogo_frutas: Esta es la forma más limpia y eficiente en Python para revisar si una clave existe dentro de un diccionario antes de intentar acceder a su valor, evitando que el programa tire un error (KeyError) si la fruta no existe.
```

```python
# 3.- Bienvenido al casino

import random

print("🎲 ¡Bienvenido al Casino! Lanzando los dados... 🎲\n")

# 1. Generar dos números aleatorios entre 1 y 6 (inclusive)
dado1 = random.randint(1, 6)
dado2 = random.randint(1, 6)

# 2. Calcular la suma de ambos dados
suma_total = dado1 + dado2

# 3. Mostrar los resultados en pantalla
print(f"Dado 1 obtenido: {dado1}")
print(f"Dado 2 obtenido: {dado2}")
print("-----------------------")
print(f"La suma de ambos dados es: {suma_total}")

# Un pequeño toque extra de casino para celebrar si sacas un 7 u 11
if suma_total in [7, 11]:
    print("\n¡Ganaste! ¡Suerte de principiante! 🏆")
elif suma_total in [2, 3, 12]:
    print("\nCraps... ¡La casa gana esta vez! 💸")
```

```python
#4. Calculadora de chocolate
# Programa: Calculadora de Porciones de Chocolate

print("🍫 ¡Bienvenido a la Calculadora de Chocolate! 🍫\n")

# 1. Solicitar los datos al usuario
# Usamos float() por si el límite o las calorías de la porción tienen decimales
limite_calorias = float(input("¿Cuál es tu límite diario de calorías? "))
calorias_por_porcion = float(input("¿Cuántas calorías tiene una porción de chocolate? "))

# 2. Validar que las calorías por porción sean mayores a cero para evitar errores matemáticos
if calorias_por_porcion <= 0:
    print("\n❌ Error: Las calorías de la porción deben ser mayores a 0.")
else:
    # 3. Calcular la cantidad máxima de porciones
    porciones_maximas = limite_calorias / calorias_por_porcion

    # 4. Mostrar el resultado en pantalla
    print("\n-----------------------")
    print(f"Resultado: Puedes comer un máximo de {porciones_maximas:.2f} porciones de chocolate.")
    print("-----------------------")

    # Un consejo extra según el resultado
    if porciones_maximas >= 3:
        print("¡Tienes bastante margen hoy! Disfrútalo con moderación. 🎉")
    elif 0 < porciones_maximas < 1:
        print("¡Cuidado! Una porción entera supera tu límite de hoy. Come solo un trocito. 🔍")
```
