2. FABRICACIION DE CERVEZA ARTESANAL
El LIC y el LSC son calculados utilizando la media y la desviación estándar de los datos del
proceso. La idea es que la mayoría de los datos del proceso se encuentren dentro de los límites
de control (entre el LIC y el LSC), y si un dato cae fuera de estos límites, se puede considerar
que el proceso está fuera de control y se requiere una acción correctiva para investigar y
corregir la causa de la variación. El valor del LIC se calcula como la media menos tres veces la
desviación estándar, y el valor del LSC se calcula como la media más tres veces la
desviación estándar. Estos límites representan el rango en el que la mayoría de los datos del
proceso deberían estar.
Para poder asegurar de que la cantidad de alcohol en cada lote de cerveza sea consistente,
debes medir el porcentaje de alcohol de cada lote y registrar los datos en una hoja de cálculo.
Esta información ya ha sido recolectada y se encuentra en el archivo alcohol_cerveza.csv.
El objetivo del ejercicio es calcular la media, la desviación estándar, el LIC y el LSC del
porcentaje de alcohol y guardarlos en variables.
Para asegurarte de que los lotes de cerveza se estén elaborando de forma consistente y que el
nivel de alcohol sea el adecuado, una vez calculadas las medidas anteriores, indicar cuáles
valores quedarían fuera de los límites de control y cuántos son.


```python
# 2. Fabricaciòn de cerveza artesanal
import numpy as np
```

```python
array = np.loadtxt("alcohol_cerveza.csv", delimiter=",", skiprows=1)
```

1.- Calcular la media y la desviacion estandar

```python
media = np.mean(array)
desviacion_estandar = np.std(array)
```

```python
media
```

```python
desviacion_estandar
```

2. Calcular los limites de control (LIC y LSC)

```python
LIC = media - 3 * desviacion_estandar
LSC = media + 3 * desviacion_estandar
```

```python
LIC
```

```python
LSC
```

3. Identificar los valores fuera de los límites


```python
#Creamos máscaras booleanas para encontrar los lotes defectuosos
```

```python
fuera_de_control = array[(array < LIC) | (array > LSC)]
cantidad_fuera = len(fuera_de_control)
```

```python
fuera_de_control
```

```python
cantidad_fuera
```

4. Mostrar los resultados en pantalla

```python
print("--- REPORTE DE CONTROL DE CALIDAD: CERVEZA ARTESANAL ---")
print(f"Media del porcentaje de alcohol: {media:.2f}%")
print(f"Desviación estándar: {desviacion_estandar:.4f}")
print(f"Límite Inferior de Control (LIC): {LIC:.2f}%")
print(f"Límite Superior de Control (LSC): {LSC:.2f}%")
print("-" * 50)
print(f"Cantidad de lotes fuera de control: {cantidad_fuera}")
if cantidad_fuera > 0:
    print(f"Valores detectados fuera de rango: {fuera_de_control}")
else:
    print("¡Excelente! Todos los lotes están dentro de los límites de control.")
```

Explicación de los puntos clave:
-np.loadtxt adaptado: Al cargar un CSV, es muy común que la primera fila sea el nombre de la columna (por ejemplo, Alcohol). Usamos skiprows=1 para ignorar ese texto y que Python no arroje un error. También manejamos el delimiter por si el archivo se separó con comas o puntos y comas.

-Cálculo de Límites: Aplicamos las fórmulas matemáticas tal cual se describen:
LIC=μ−3σ
LSC=μ+3σ

-Filtrado Condicional: Para saber cuáles están fuera, usamos el operador lógico | (que significa "O" en NumPy). Le decimos al programa: "Búscame los valores que sean menores al LIC O mayores al LSC".

5. Filtra el array utilizando np.logical_or(condicion1, condicion2) para obtener los valores
fuera de los límites de control.

a) definimos las condiciones

```python
condicion1 = array < LIC
condicion2 = array > LSC
```

b) Filtramos el array usando np.logical_or

```python
valores_fuera = array[np.logical_or(condicion1, condicion2)]
```

c) Contamos cuántos valores quedaron fuera

```python
cantidad_fuera = len(valores_fuera)
```

Muestra de los resultados

```python
print(f"Cantidad de lotes fuera de control: {cantidad_fuera}")
print(f"Valores de los lotes fuera de control: {valores_fuera}")
```

Esta función compara elemento por elemento del array y devuelve True (verdadero) si el dato cumple con la condicion1 (es más bajo que el límite inferior) O si cumple con la condicion2 (es más alto que el límite superior).

Si no cumple ninguna de las dos (es decir, el nivel de alcohol es el adecuado y está dentro del rango), devuelve False. Al final, NumPy usa esas respuestas para dejar pasar solo los lotes "defectuosos".
