```python
# 1 a) Usando la siguiente lista imprime el quinto elemento
#  lista = ["manzana", "pera", "naranja", "plátano", "kiwi", "uva", "piña", "mango", "sandía", "melón"]
lista = ["manzana", "pera", "naranja", "plátano", "kiwi", "uva", "piña", "mango", "sandía", "melón"]
# Imprimimos el quinto elemento el cual en python corresponde al indice 4 ya que el conteo comienza desde 0
print(lista[4])
```

```python
# 1 b) Imprime del cuarto al sexto elemento de la lista
# El 4° elemento ("plátano") tiene el índice 3.
# El slicing [inicio:fin] incluye el índice de inicio pero excluye el de fin, así que debemos llegar hasta el índice 6 para que tome el 3, 4 y 5.
# Slicing desde el ìndice 3 hasta el 6 (no incluido)

Resultado = lista[3:6]
print(Resultado)
```

```python
# 1 c) Elimina el elemento en la posiciòn 3
# Eliminamos el elemento en la posicion 3 correspondiente al ìndice 2
del lista[2]
print(lista)
```

```python
# 1 d) Agrega dos nuevos elementos a la lista
lista = ["manzana", "pera", "naranja", "plátano", "kiwi", "uva", "piña", "mango", "sandía", "melón"]
# Utilizamos append para añadir uno por uno cada elemento al final de la lista
lista.append("fresa")
lista.append("arándano")
print(lista)
```

```python
# 1 e) Ordena la lista alfabèticamente
lista = ["manzana", "pera", "naranja", "plátano", "kiwi", "uva", "piña", "mango", "sandía", "melón"]
# Utilizamos el comando sort para ordenar alfabèticamente la lista
lista.sort()
print(lista)
```

```python
# 1. f) Genera un diccionario usando lista1 como llaves y lista como valores.
# lista = ["manzana", "pera", "naranja", "plátano", "kiwi", "uva", "piña", "mango", "sandía", "melón"]
# lista1 = ["uno", "dos", "tres", "cuatro", "cinco", "seis", "siete", "ocho", "nueve", "diez"]
# zip() empareja los elementos posición por posición
# dict() los convierte en clave: valor
lista = ["manzana", "pera", "naranja", "plátano", "kiwi", "uva", "piña", "mango", "sandía", "melón"]
lista1 = ["uno", "dos", "tres", "cuatro", "cinco", "seis", "siete", "ocho", "nueve", "diez"]
diccionario = dict(zip(lista1, lista))
print(diccionario)
```
