● El IMC es una medida importante que nos ayuda a determinar si una persona
tiene un peso saludable para poder ser un astronauta. Para calcularlo se utiliza la
fórmula
IMC = m / h*h
Donde m corresponde al peso del astronauta en kilogramos, y h su estatura en metros
El rendimiento en las diferentes evaluaciones permitirá también aceptar o rechazar a los
postulantes. Para ello:


1. Crea un DataFrame con el archivo entregado.

```python
import pandas as pd
```

## #Crear el DataFrame a partir del archivo CSV entregado

```python
df_astronautas = pd.read_csv('resumen_resultados_astronautas.csv')
```

# #Mostrar las primeras filas para verificar la lectura correcta

```python
print(df_astronautas.head())
```

2. Crea la columna "IMC" usando la fórmula entregada.

```python
df_astronautas["altura"] = df_astronautas["altura"] / 100
```

```python
df_astronautas["IMC"] = df_astronautas["peso"] / (df_astronautas["altura"] ** 2)
```

```python
df_astronautas.head()
```

3. Crea la columna "promedio_evaluaciones" con el promedio de las 4 columnas
de evaluaciones. Si el IMC está entre 18.5 y 24.9, entonces el candidato
tendrá un IMC adecuado para ser astronauta. Filtra el DataFrame para
obtener solo aquellas filas donde el IMC esté dentro de los límites
entregados.

# 3.1. Crear la columna "promedio_evaluaciones"

```python
columnas_evaluaciones = [
    "evaluacion_fisica",
    "evaluacion_psicologica",
    "evaluacion_adaptativa",
    "evaluacion_medica",
]
```

# Calculamos el promedio de esas 4 columnas a lo largo de las filas (axis=1)

```python
df_astronautas["promedio_evaluaciones"] = df_astronautas[columnas_evaluaciones].mean(axis=1)
```

# 3.2. Filtrar el DataFrame según el rango de IMC adecuado (entre 18.5 y 24.9)
# Usamos .between() que incluye ambos límites por defecto

```python
df_astronautas_filtrado = df_astronautas[df_astronautas["IMC"].between(18.5, 24.9)]
```

# Mostrar los candidatos preseleccionados

```python
print(f"Cantidad de candidatos con IMC adecuado: {len(df_astronautas_filtrado)}")
df_astronautas_filtrado.head()
```

4. El promedio de las 4 evaluaciones debe ser mayor o igual a 87. Filtra el
DataFrame resultante y agrega esta condición al filtro aplicado en el paso
anterior.

# 4. Combinar ambas condiciones en un solo filtro
# Condición 1: IMC entre 18.5 y 24.9
# Condición 2: promedio_evaluaciones mayor o igual a 87

```python
df_astronautas_final = df_astronautas[(df_astronautas["IMC"].between(18.5, 24.9)) & (df_astronautas["promedio_evaluaciones"] >= 87)]
```

```python
print(f"Cantidad de candidatos que cumplen todos los requisitos: {len(df_astronautas_final)}")
```

```python
df_astronautas_final.head()
```

5. Reinicia el índice del DataFrame resultante.

```python
# Reiniciar el índice del DataFrame resultante
df_astronautas_final = df_astronautas_final.reset_index(drop=True)

# Mostrar el DataFrame con el nuevo índice ordenado
df_astronautas_final.head()
```

```python
# 6. Modificar o crear la columna "califica" con el valor "Si"
df_astronautas_final["califica"] = "Si"

# Mostrar el DataFrame final para verificar el cambio
df_astronautas_final.head()
```

```python
# 7. Exportar el DataFrame resultante a un archivo CSV separado por comas
df_astronautas_final.to_csv("astronautas_calificados.csv", index=False, sep=",")

print("¡El archivo 'astronautas_calificados.csv' ha sido creado con éxito!")
```
