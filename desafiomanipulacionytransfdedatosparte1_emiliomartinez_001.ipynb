```python
# a) Carga los datos y crea un DataFrame con cada uno de ellos.
```

```python
import pandas as pd
df_incidents = pd.read_pickle('incidents.pkl')
df_officers = pd.read_pickle('officers.pkl')
df_subjects = pd.read_pickle('subjects.pkl')

print("--- Formato inicial de las tablas ---")
print(f"Incidentes: {df_incidents.shape}")
print(f"Oficiales: {df_officers.shape}")
print(f"Sujetos: {df_subjects.shape}\n")
```

```python
# b) Genera una tabla que contenga la unión de las 3 tablas. hint: utiliza sufijos para para las
# columnas que se llaman igual usando el parámetro suffixes de pd.merge().
```

```python
# Primero unimos incidentes con oficiales. Asumimos que se unen por una columna común (ej. 'case_number').
# Usamos suffixes para diferenciar columnas con el mismo nombre (como 'gender' o 'race')
```

```python
df_merge_1 = pd.merge(df_incidents, df_officers, on='case_number', suffixes=('_incident', '_officer'))
```

```python
df_merge_1
```

```python
# Luego unimos el resultado con la tabla de sujetos.
```

```python
df_final = pd.merge(df_merge_1, df_subjects, on='case_number', suffixes=('', '_subject'))
```

```python
df_final
```

```python
# c. Verifica si hay filas duplicadas; si es así, elimínalas.
```

```python
duplicados = df_final.duplicated().sum()
print(f"Cantidad de filas duplicadas encontradas: {duplicados}")

if duplicados > 0:
    df_final = df_final.drop_duplicates()
    print("Filas duplicadas eliminadas exitosamente.\n")
else:
    print("No se encontraron filas duplicadas.\n")
```

```python
# d. A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo
# y Máximo. (hint: Utiliza el método apply sobre la columna.)
```

```python
df_jobs = pd.read_csv("Cleaned_DS_Jobs.csv")
```

```python
print(df_jobs.head())
```

```python
import pandas as pd
import numpy as np

# 1. Definimos una función para limpiar y extraer los salarios
def extraer_limite_salario(texto_salario, tipo="minimo"):
    # Asegurarnos de que el valor sea un string y no esté vacío
    if pd.isna(texto_salario) or not isinstance(texto_salario, str):
        return np.nan

    # Limpieza inicial: quitar el signo de dólar y pasar a minúsculas
    texto_limpio = texto_salario.replace('$', '').lower()

    # Separar el rango por el guion '-'
    if '-' in texto_limpio:
        partes = texto_limpio.split('-')

        # Elegir la parte izquierda para el mínimo o la derecha para el máximo
        if tipo == "minimo":
            valor_str = partes[0].strip()
        else:
            # A veces el lado máximo tiene texto extra como "(Glassdoor est.)", nos quedamos solo con el primer bloque
            valor_str = partes[1].strip().split()[0]

        # Controlar si el formato usa "k" (miles) o el número completo
        if 'k' in valor_str:
            valor_str = valor_str.replace('k', '')
            return float(valor_str) * 1000
        else:
            # Quitar comas si los números son del estilo 100,000
            valor_str = valor_str.replace(',', '')
            return float(valor_str)

    return np.nan

# 2. Aplicar la función usando .apply() y lambdas para crear las nuevas columnas
df_jobs["Salario Estimado Mínimo"] = df_jobs["Salary Estimate"].apply(lambda x: extraer_limite_salario(x, tipo="minimo"))
df_jobs["Salario Estimado Máximo"] = df_jobs["Salary Estimate"].apply(lambda x: extraer_limite_salario(x, tipo="maximo"))

# 3. Verificar el resultado impreso en pantalla
print(df_jobs[["Salary Estimate", "Salario Estimado Mínimo", "Salario Estimado Máximo"]].head())
```

```python
# e. Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint:
# utilice reemplazo con diccionario usando el método replace sobre la columna.)
```

```python
# Size Descripción
10000+ employees Mega Empresas
5001 to 10000 employees Grandes Empresas
1001 to 5000 employees Medianas Empresas
201 to 500 employees Pequeñas Empresas
51 to 200 employees Pequeñas Grandes Empresas
501 to 1000 employees Microempresas
Unknown Empresas sin Información
```

```python
# 1. Crear el diccionario con el mapeo de la tabla proporcionada
diccionario_size = {
    "10000+ employees": "Mega Empresas",
    "5001 to 10000 employees": "Grandes Empresas",
    "1001 to 5000 employees": "Medianas Empresas",
    "201 to 500 employees": "Pequeñas Empresas",
    "51 to 200 employees": "Pequeñas Grandes Empresas", # Nota: así venía en tu tabla
    "501 to 1000 employees": "Microempresas",
    "Unknown": "Empresas sin Información"
}

# 2. Aplicar el reemplazo utilizando el método .replace() sobre la columna Size
df_jobs["Size"] = df_jobs["Size"].replace(diccionario_size)

# 3. Verificar los cambios imprimiendo el conteo de los nuevos valores
print("Valores recodificados en la columna 'Size':")
print(df_jobs["Size"].value_counts())
```
